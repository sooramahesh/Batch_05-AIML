```python
from collections import deque

def bfs(graph, start):
    visited = set()
    queue = deque([start])

    while queue:
        current = queue.popleft()

        if current not in visited:
            print(current, end=" ")
            visited.add(current)

            queue.extend(neighbor for neighbor in graph[current] if neighbor not in visited)

# Example usage:
# Representing the graph using an adjacency list
graph = {
    0: [1, 3],
    1: [0, 2],
    2: [1,4,5],
    3: [0,4],
    4: [3,2,5],
    5: [2,4]
}

start_vertex = 0
print(f"BFS traversal starting from vertex {start_vertex}:")
bfs(graph, start_vertex)
```
